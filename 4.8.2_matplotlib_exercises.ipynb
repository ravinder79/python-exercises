```python
x = range(-100,100)


y = [(x**2 - x + 2) for x in x ]

plt.plot(x,y)
plt.title('$y = x2 -x + 2$')
plt.annotate('Origin', xy=(0, 0), xytext=(0, 2500),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
import math
#x1 = range(0,10)
x1 = [i* 3.14/32 for i in range (0,100)]
x = [i*1 for i in range(0,10)]
y1 = [math.sqrt(x) for x in x]
y2 = [x**3 for x in x1]
y3 = [math.tan(x) for x in x1]
y4 = [2**x for x in x1]
plt.figure(figsize=(16, 9))

plt.plot(x,y1, label = '$sqrt(x)$')
plt.plot(x1,y2, label = '$x^3$')
plt.plot(x1,y3, label = '$tan(x)$')
plt.plot(x1,y4, label = '$2^x$')
plt.legend(loc='best')
plt.ylim(-10,10)
```
