```python
x = range(-100,100)


y = [(x**2 - x + 2) for x in x ]

plt.plot(x,y)
plt.title('$y = x2 -x + 2$')
plt.annotate('Origin', xy=(0, 0), xytext=(0, 2500),
             arrowprops={'facecolor': 'blue'})
plt.xlabel("x")
plt.ylabel("y", rotation = 0)
plt.show()
```

```python
import numpy as np
x = np.linspace(-2 * math.pi, 2 * math.pi, 1000)
#x = [i* math.pi/32 for i in range (-120,120) ]
y = np.tan(x)
#y3 = [math.tan(x1) for x1 in x1]
y[:-1][np.diff(y)<0] = np.nan
plt.xlabel("x")
plt.grid()
plt.ylabel("$tan(x)$", rotation = 0)
plt.title('$y = tan(x)$')
radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']
plt.xticks(radians, radian_labels)
plt.plot(x,y)
plt.ylim(-7, 7)
plt.xlim(-2*math.pi, 2*math.pi)
```
