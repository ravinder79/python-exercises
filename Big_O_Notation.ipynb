```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

y = 0x + 1 and label the curve "O(1)"

```python
x = range(0,20)
y = [0*x +1 for x in x]
plt.figure(figsize=(14, 10))
plt.plot(x,y, label = '0(1)')
plt.ylim = (0, 2)
plt.xlim = (0, 20)

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')

x = [i*0.1 for i in range(1,150)]
y = [math.log(x) for x in x]
plt.plot(x,y, label = 'i(log n)')
plt.legend(loc='best')

x = range(0,20)
y = x
plt.plot(x,y, label = '0(n)')
plt.title("0(n)")
plt.legend(loc='best')

x = [i for i in range(1,20)]
y = [math.log(x)*x for x in x]
plt.plot(x,y, label = '0(n log n)')
plt.legend(loc='best')

x = [i for i in range(1,10)]
y = [x**2 for x in x]
plt.plot(x,y, label = '0(n^2)')
plt.legend(loc='best')

x = [i for i in range(0,6)]
y = [2**x for x in x]
plt.plot(x,y, label = '0(2^n)')
plt.legend(loc='best')

x = [i for i in range(0,6)]
y = [math.factorial(x) for x in x]
plt.plot(x,y, label = '0(n!)')
plt.legend(loc='best')

x = [i for i in range(0,4)]
y = [(x**x) for x in x]
plt.plot(x,y, label = '0(n^n)')
plt.legend(loc='best')
plt.grid()
plt.ylim = (0,100)
plt.title("Big O Notation")
```

$y = log(x)$ and label the curve "O(log n)"

```python
x = [i*0.1 for i in range(1,100)]
y = [math.log(x) for x in x]
plt.plot(x,y, label = 'i(log n)')
plt.title("0(log n)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.grid()
plt.legend(loc='best')
```

$y = x$ and label the curve "O(n)"

```python
x = range(0,20)
y = x
plt.plot(x,y, label = '0(n)')
plt.title("0(n)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.grid()
```

$y = x * log(x)$ and label it "O(n log n)"

```python
x = [i*0.1 for i in range(1,100)]
y = [math.log(x)*x for x in x]
plt.plot(x,y, label = '0(n log n)')
plt.title("0(n log n)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.grid()
```

$y = x^2$ and label it "O(n^2)"

```python
x = [i for i in range(0,10)]
y = [x**2 for x in x]
plt.plot(x,y, label = '0(n^2)')
plt.title("0(n^2)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.grid()
```

$y = 2^x$ and label it "O(2^n)"

```python
x = [i for i in range(0,10)]
y = [2**x for x in x]
plt.plot(x,y, label = '0(2^n)')
plt.title("0(2^n)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.grid()
```

$y = n!$ and label is "O(n!)"

```python
x = [i for i in range(0,5)]
y = [math.factorial(x) for x in x]
plt.plot(x,y, label = '0(n!)')
plt.title("0(n!)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')
plt.grid()
```

$y = n^n$ and label it "O(n^n)"

```python
x = [i for i in range(0,5)]
y = [(x**x) for x in x]
plt.title("0(n^n)")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.grid()
plt.plot(x,y, label = '0(n^n)')
plt.legend(loc='best')
plt.xlim = (0,5)
plt.ylim = (0,250)
plt.show()
```
